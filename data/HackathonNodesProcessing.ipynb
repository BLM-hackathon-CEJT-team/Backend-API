Converted the API .js files into JSON to import into python/pandas

```python
dead_json = open(fileHad,mode='rt')
alive_json = open(fileWanted, "wt")
```

These string replacements were variant per .js file, and this is just the last iteration. There were ~6 so it was worth doing by hand.

```python
for line in dead_json:
    alive_json.write(line.replace(r',',r',"').replace(r':',r'":').replace(r'board',r'"board').replace(r'{', r'"X":{').replace(r'},"',r'},'))
    
alive_json.close()
dead_json.close()
```
